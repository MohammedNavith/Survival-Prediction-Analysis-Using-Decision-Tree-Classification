```python
from sklearn.preprocessing import LabelEncoder
```

```python
df1 = df.copy()
```

```python
object1 = LabelEncoder()
```

```python
df1['T_Stage_1'] = object1.fit_transform(df1['T Stage '])
df1['Estrogen_Status_1'] = object1.fit_transform(df1['Estrogen Status'])
df1['Survival_Status_1'] = object1.fit_transform(df1['Status'])
```

```python
df1.head()
```

```python
df1.columns
```

```python
numerical_cols = ['Age', 'Tumor Size', 'Estrogen_Status_1', 'T_Stage_1', 'Survival Months']
correlation_matrix = df1[numerical_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

Key Observations:
Diagonal Values (Correlation = 1):

These represent the correlation of a variable with itself, which is always 1.
Tumor Size and T Stage:

A strong positive correlation of 0.81 indicates that as the tumor size increases, the T stage tends to be higher. This makes sense clinically, as T stage often reflects tumor size.
Estrogen Status and Survival Months:

A weak positive correlation of 0.13 suggests that patients with positive estrogen status might survive slightly longer, but the relationship is not very strong.
Tumor Size and Survival Months:

A slight negative correlation of -0.09 indicates that larger tumors may be associated with shorter survival, but the effect is minimal.
Age and Other Variables:

Age shows very weak correlations with all variables, suggesting that age alone may not be a significant factor in determining these outcomes.
T Stage and Survival Months:

A weak negative correlation of -0.09 shows that higher T stages may slightly reduce survival months, though the relationship is not strong.

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
x = df1[['Age','T_Stage_1','Estrogen_Status_1','Tumor Size','Regional Node Examined','Reginol Node Positive']]
y = df1['Survival_Status_1']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 10)
```

```python
dtf = DecisionTreeClassifier()
```

```python
dtf.fit(x_train,y_train)
```

```python
y_pred = dtf.predict(x_test)
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print("Confusion Matrix:")
print(conf_matrix)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("Confusion Matrix:")
print(conf_matrix)
```
